```python
#  1. Budget COGS (Cost of Goods Sold)
# This is the estimated cost of producing the goods you expect to sell.

# Calculation:
# 800 units Ã— â‚¹100 = â‚¹80,000
# âœ… Budget COGS = â‚¹80,000

# ðŸ“˜ 2. Budget Sales
# This is the expected revenue from selling 800 units.

# Calculation:
# 800 units Ã— â‚¹150 = â‚¹1,20,000
# âœ… Budget Sales = â‚¹1,20,000

# ðŸ“˜ 3. Budget Margin
# This is the expected gross profit (before other expenses).

# Calculation:
# Budget Sales âˆ’ Budget COGS
# â‚¹1,20,000 âˆ’ â‚¹80,000 = â‚¹40,000
# âœ… Budget Margin = â‚¹40,000

# ðŸ“˜ 4. Budget Profit
# This is the expected net profit after subtracting all costs, including fixed expenses.

# Calculation:
# Budget Margin âˆ’ Other expenses
# â‚¹40,000 âˆ’ â‚¹5,000 = â‚¹35,000
# âœ… Budget Profit = â‚¹35,000

# Now suppose actual results differ slightly:

# You actually sold 850 units

# Actual cost per unit = â‚¹105

# Actual selling price = â‚¹145

# Other expenses = â‚¹6,000

# You still produced 1,000 units

# ðŸ“˜ 5. COGS (Actual)
# Actual cost of goods you sold, not produced.

# Calculation:
# 850 units Ã— â‚¹105 = â‚¹89,250
# âœ… COGS = â‚¹89,250

# ðŸ“˜ 6. Inventory
# You produced 1,000 units, sold 850 â†’ 150 units remain in stock.

# Calculation (at cost):
# 150 units Ã— â‚¹105 = â‚¹15,750
# âœ… Inventory = â‚¹15,750

# ðŸ“˜ 7. Margin (Actual)
# This is your actual gross profit.

# Calculation:
# 850 units Ã— â‚¹145 = â‚¹1,23,250 (actual sales)
# Margin = Sales âˆ’ COGS = â‚¹1,23,250 âˆ’ â‚¹89,250 = â‚¹34,000
# âœ… Margin = â‚¹34,000

```
